```python
import cv2
import numpy as np
import face_recognition
from keras_preprocessing import image
from tensorflow.keras.models import model_from_json
```

```python
webcam_vedeo_stream = cv2.VideoCapture(0)
face_exp_model = model_from_json(open("model/facial_expression_model_structure.json",'r').read())
face_exp_model.load_weights("model/facial_expression_model_weights.h5")
emotion_labels = ('angry','dsigust','fear','happy','sad','surprise','neutral')

all_face_locations = []
```

```python

while True:
    ret,current_frame   = webcam_vedeo_stream.read()
    current_frame_small = cv2.resize(current_frame, (0,0),fx = 0.25,fy = 0.25)
    all_face_location   = face_recognition.face_locations(current_frame_small,model='hog')
    
    for index, current_face_location in enumerate(all_face_location):
        top_pos,right_pos,bottom_pos,left_pos = current_face_location
        
        top_pos    = top_pos*4
        right_pos  = right_pos*4
        bottom_pos = bottom_pos*4
        left_pos   = left_pos*4
        print("found face {} at top: {} ,right: {} ,bottom: {} ,left: {}".format(index+1,top_pos,right_pos,bottom_pos,left_pos))

        cv2.rectangle(current_frame,(left_pos,top_pos),(right_pos,bottom_pos),(0,255,0),2)
        current_face_iamge = current_frame[top_pos:bottom_pos,left_pos:right_pos]
        
        current_face_iamge = cv2.cvtColor(current_face_iamge,cv2.COLOR_RGB2GRAY)
        current_face_iamge = cv2.resize(current_face_iamge,(50,50))
        img_pixel = image.img_to_array(current_face_iamge)
        img_pixel = np.expand_dims(img_pixel,axis=0)
        img_pixel /=255
        
        exp_predictions = face_exp_model(img_pixel)
        max_index = np.argmax(exp_predictions[0])
        emotion_label = emotion_labels[max_index]
        
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(current_frame,emotion_label,(left_pos,top_pos-10), font,0.6,(153,255,221),1)
    cv2.imshow("Webcom video", current_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
 
webcam_video_stream.release()
cv2.destroyAllWindows()
```
